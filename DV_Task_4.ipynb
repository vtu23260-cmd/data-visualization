```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import parallel_coordinates

# Load dataset
df = pd.read_csv("/content/sigma_cabs.csv")

# Select numeric variables for multivariate analysis
numeric_cols = [
    'Trip_Distance', 'Customer_Since_Months', 'Life_Style_Index',
    'Customer_Rating', 'Var1', 'Var2', 'Var3'
]

# Drop NaN rows for clean plotting
df_clean = df[numeric_cols + ['Type_of_Cab']].dropna()

# 1. Scatterplot Matrix
sns.pairplot(df_clean[numeric_cols])
plt.suptitle("Scatterplot Matrix", y=1.02)
plt.show()

# 2. Parallel Coordinates Plot
plt.figure(figsize=(10, 6))
parallel_coordinates(df_clean, 'Type_of_Cab',
                     color=['#FF6347', '#4682B4', '#3CB371'])
plt.title("Parallel Coordinates Plot")
plt.xticks(rotation=45)
plt.show()

# 3. Line Graph
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_clean, x='Customer_Since_Months', y='Trip_Distance',
             hue='Type_of_Cab', ci=None)
plt.title("Line Graph: Trip Distance vs Customer Since Months")
plt.show()

# 4. Stacked Bar Chart
mean_values = df_clean.groupby('Type_of_Cab')[numeric_cols].mean()
mean_values.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title("Stacked Bar Chart: Mean Values per Cab Type")
plt.ylabel("Mean Values")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("sigma_cabs.csv")

# Drop rows with missing values for selected columns
df_uni = df.dropna(subset=["Type_of_Cab", "Gender", "Trip_Distance", "Customer_Rating", "Life_Style_Index"])

# Set plot style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# -------------------------------
# CATEGORICAL DATA VISUALIZATION
# -------------------------------

# 1. Bar Chart - Type of Cab
sns.countplot(data=df_uni, x="Type_of_Cab")
plt.title("Bar Chart: Type of Cab")
plt.tight_layout()
plt.show()

# 2. Pie Chart - Gender
gender_counts = df_uni["Gender"].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart: Gender Distribution")
plt.axis('equal')
plt.tight_layout()
plt.show()

# ---------------------------------
# CONTINUOUS DATA VISUALIZATION
# ---------------------------------

# 3. Scatterplot: Trip Distance vs Index
plt.scatter(df_uni.index, df_uni["Trip_Distance"], alpha=0.5)
plt.title("Scatterplot: Trip Distance vs Index")
plt.xlabel("Index")
plt.ylabel("Trip Distance")
plt.tight_layout()
plt.show()

# 4. Line Plot - Trip Distance over Index
sns.lineplot(x=df_uni.index, y="Trip_Distance", data=df_uni)
plt.title("Line Plot: Trip Distance over Index")
plt.tight_layout()
plt.show()

# 5. Strip Plot - Customer Rating by Gender
sns.stripplot(data=df_uni, x="Gender", y="Customer_Rating", jitter=True)
plt.title("Strip Plot: Customer Rating by Gender")
plt.tight_layout()
plt.show()

# 6. Swarm Plot - Life Style Index by Cab Type
sns.swarmplot(data=df_uni, x="Type_of_Cab", y="Life_Style_Index", size=3)
plt.title("Swarm Plot: Life Style Index by Cab Type")
plt.tight_layout()
plt.show()

# 7. Histogram - Trip Distance
sns.histplot(df_uni["Trip_Distance"], bins=30, kde=False)
plt.title("Histogram: Trip Distance")
plt.tight_layout()
plt.show()

# 8. Density Plot - Trip Distance
sns.kdeplot(df_uni["Trip_Distance"], shade=True)
plt.title("Density Plot: Trip Distance")
plt.tight_layout()
plt.show()

# 9. Rug Plot - Customer Rating
sns.rugplot(df_uni["Customer_Rating"])
plt.title("Rug Plot: Customer Rating")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Load dataset
file_path = "/content/sigma_cabs.csv"
df = pd.read_csv(file_path)

# Replace NaN in hierarchy columns with "Unknown"
df[["Type_of_Cab", "Destination_Type", "Gender"]] = df[["Type_of_Cab", "Destination_Type", "Gender"]].fillna("Unknown")

# --- 5a) TreeMap ---
fig_treemap = px.treemap(
    df,
    path=["Type_of_Cab", "Destination_Type", "Gender"],  # hierarchy
    values="Trip_Distance",                               # size of boxes
    color="Customer_Rating",                             # coloring by rating
    color_continuous_scale="Viridis",
    title="TreeMap of Cab Trips by Type, Destination, and Gender"
)
fig_treemap.show()

# --- 5b) Sunburst ---
fig_sunburst = px.sunburst(
    df,
    path=["Type_of_Cab", "Destination_Type", "Gender"],  # hierarchy
    values="Trip_Distance",                               # size of segments
    color="Customer_Rating",                             # coloring by rating
    color_continuous_scale="Plasma",
    title="Sunburst of Cab Trips by Type, Destination, and Gender"
)
fig_sunburst.show()
```
